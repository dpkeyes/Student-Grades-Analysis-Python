```python
# Import dependencies
import pandas as pd
import numpy as np

# Load in csv files
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Convert csv files into pandas data frames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)
```

```python
# Join the school data and student data tables on the 'school_name' column
full_data = pd.merge(student_data, school_data, how="left", on="school_name")
full_data.head()
```

```python
# Rename column titles and rearrange column order to be more natural
full_data = full_data.rename(columns={"student_name": "Student Name", "gender": "Gender", "grade": "Grade", "school_name": "School Name", "reading_score": "Reading Score", "math_score": "Math Score", "type": "School Type", "size": "School Size", "budget": "School Budget"})
full_data = full_data[["Student ID", "Student Name", "Gender", "Grade", "Reading Score", "Math Score", "School ID", "School Name", "School Type", "School Size", "School Budget"]]
full_data.head()
```
